```python
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.corpus import stopwords
nltk.download('wordnet')
import re
```

```python
df = pd.read_csv('E:/vs projects/NLP Section/data/tripadvisor_hotel_reviews.csv')
df.head()
```

```python
df.info()
```

```python
df['Review'][0]
```

```python
df['review lower'] = df['Review'].str.lower()
df.head()
```

```python
en_stopwords = stopwords.words('English')
en_stopwords.remove("not")
```

```python
df['review_no_stopwords'] = df['review lower'].apply(lambda x: ' '.join(word for word in x.split() if word not in (en_stopwords)))
df['review_no_stopwords'][0]
```

```python
df['review_no_stopwords_no_punc'] = df.apply(lambda x: re.sub(r"[*]","star",x['review_no_stopwords']), axis=1)
df.head()
```

```python
df['review_no_stopwords_no_punc'] = df.apply(lambda x: re.sub(r"[^\w\s]","",x['review_no_stopwords_no_punc']), axis=1)
df.head()
```

```python
df['tokenized'] = df.apply(lambda x: word_tokenize(x['review_no_stopwords_no_punc']), axis=1)
df['tokenized'][0]
```

```python
ps = PorterStemmer()
```

```python
df['stemmed'] = df['tokenized'].apply(lambda tokens: [ps.stem(token) for token in tokens])
df.head()
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
df['lemmatized'] = df['tokenized'].apply(lambda tokens: [lemmatizer.lemmatize(token) for token in tokens])
df['lemmatized'][0]
```

```python
tokens_clean = sum(df['lemmatized'], [])
```

```python
unigrams = (pd.Series(nltk.ngrams(tokens_clean, 1)).value_counts())
print(unigrams)
```

```python
bigrams = (pd.Series(nltk.ngrams(tokens_clean, 2)).value_counts())
print(bigrams)
```
